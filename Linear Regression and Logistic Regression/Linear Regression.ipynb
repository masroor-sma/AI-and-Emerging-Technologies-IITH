### Generating data 'X' ranges between -10 to 10 
### 'Y' is a linear combination of parameters as Y = ax+b+gaussian(0,epsilon)

```python
import numpy as np                             #Mathematical Library
```

```python
Number_of_points = 1000                        #Sample points
X_data = np.linspace(-10,10,Number_of_points)  #Values betwn -10 and 10 (1000 samples)
W0_true = 11                                   #y = ax + b (a =W1 ,b = W0 with actual values)
W1_true = 4.5
Y_data = W1_true*X_data + W0_true + np.random.normal(0,0.5,Number_of_points) #Random sampling with Mean = 0 and SD = 0.5
```

### Split the data into training and testing set

```python
Number_of_training = int(Number_of_points*0.8)  # 80% training data points
Number_of_testing = int(Number_of_points*0.2)   # 20% Testing data points
index = np.random.permutation(Number_of_points) # Shuffling the points to avoid overfitting and underfitting


#index = np.random.permutation(10)              # We take random permutation for shuffling the data
#print(index)                                   # Used to print the values
#print(index[:4])
#A random permutation of 1..N is generated


```

```python
X_train = X_data[index[:Number_of_training]]   # Samples from start to Number of Training
Y_train = Y_data[index[:Number_of_training]]   # Samples from Number of training to End (: read)
```

```python
X_test = X_data[index[Number_of_training:]]    #Same as Training
Y_test = Y_data[index[Number_of_training:]]
```

### Plot the data points

```python
import matplotlib.pyplot as plt
def plot(x,y,title):
    fig,(ax1) = plt.subplots(1)
    if title!='Loss':
        ax1.plot(x,y,'x')
    else:
        ax1.plot(x,y)
        
    ax1.set_title(title, fontdict={'fontsize': 15, 'fontweight': 'medium'})
def plot_curve(x,y,W0,W1,title):
    fig,(ax1) = plt.subplots(1)
    #ax1.plot(x,y,'x')
    ax1.set_title(title, fontdict={'fontsize': 15, 'fontweight': 'medium'})
    y = W0 + W1*x
    #ax1.plot(x,y,'o')
    ax1.plot([X_data[0],X_data[-1]],[W0_true+W1_true*X_data[0],W0_true+W1_true*X_data[-1]],color = 'blue')
    ax1.plot([X_data[0],X_data[-1]],[W0+W1*X_data[0],W0+W1*X_data[-1]],color = 'red')

    
plot(X_train[0:10],Y_train[0:10],'Few data points')   #Plotting the 10 points
```

### Fitting a 1st order polynomial  of type y = W0 + W1*x. 
We want to learn the parameters W0 and W1. Lets initialize W0 and W1 with random values

```python

def compute_loss(x,y,W0,W1):
    return np.sum(np.power(W0+W1*x - y,2))/(2.0*Number_of_training)   #Sum-->Squaring the formula y=mx+c
    #Here summation is there so we are taking average
```

### Learning the parameters of the model for some 'N' epochs, by minimizing squared-difference error
$L(W) = \frac{1}{2}\Sigma_{n=1}^{N}(W^Tx_i - y_i)^2$                                                            

L = Loss equation

MSE(Minimized value, Should be less) = (Predicted - Actual)*2



```python
Epochs = 1000                                                      #We are learning values of W 1000 times (!000 times learning is done)
learning_rate = 0.001                                              #Learning rate
W0 = np.random.normal(0,0.2,1)                                     #Best approximate the loss
W1 = np.random.normal(0,0.2,1)                                     #Iteration goes for the 1000 times and it keeps correcting itself
Loss = []
for epoch in range(Epochs):
    dW0 = 0;dW1 = 0
    for n in range(Number_of_training):
        dW0 = dW0 + (W0 + W1*X_train[n] - Y_train[n])*1            #gradience calculation(Intially it was 0)   W0 + W1*X_train[n](Actual) - Y_train[n](Pred)
        dW1 = dW1 + (W0 + W1*X_train[n] - Y_train[n])*X_train[n]   #Gradiance cal
    W0 = W0 - learning_rate*dW0/Number_of_training                 #How it is improving the line
    W1 = W1 - learning_rate*dW1/Number_of_training                 #Average 
    Loss.append(compute_loss(X_train,Y_train,W0,W1))               # W = W1 + W0 For loss computing we need W0 and W1 values
    if epoch%50==0:                                                #We are calculating the W1 and W0
        plot_curve(X_train,Y_train,W0,W1,'Training')
        print("Epoch : "+str(epoch)+"\\"+str(Epochs) + " Train loss: "+str(compute_loss(X_train,Y_train,W0,W1)))
            #plot_curve(X_train,Y_train,W,'Training')


plot(range(len(Loss)),Loss,'Loss')



#26 mins of the lecture
#W = W1 + W0    
        
    
        
```
