<h4> <b> B. Implementation of Expectation and Maximization (EM) algorithm : </b> This part focuses on implementing the EM algorithm from scratch to estimate the parameters of the Gaussian Mixture Model (GMM).   </h4>

<dt> <h4> 1. Chose the number of mixtures (M) and initialize the parameters of GMM model </h4> </dt>
<dd> <h4> - Make sure that the diagonal elements of covariance are positive, and the weights of Gaussians are positive and sum to 1 for satisfying the axioms of probability. </h4> </dd>

<dt> <h4>  2. Expectation step :  Evaluate the responsibility of Gaussians in generating the data points using the current estimated parameters i.e $\hat\mu_{k}$ , $\hat\Sigma_{k}$ and $\hat\pi_{k}$, here k indicates the Gaussian in GMM </h4> </dt>
<br>

<dd> <h4> <center> $\gamma_{nk} = \frac{\pi_{k} N ( \textbf{x}_{n} / \mu_{k} , \Sigma_{k} )}{ \sum_{m=1}^{M} \pi_{m} N ( x_{n} / \mu_{m} , \Sigma_{m} ) }    \;\;\;\;\;\; âˆ€ \;\;k=1,2,...,M \;\; and \;\; n=1,2,...,N$  </center> </h4> </dd>

<br>

<dd> <h4>  Where N is the total number of data points and M is the total number of Gaussians in GMM  </h4> </dd>

<dd> <h4>  - $\gamma_{nk}$ represents the responsibility of $k^{th}$ Gaussian in generating the $n^{th}$ data point  </h4> </dd>

<dd> <h4>  - $\sum_{n=1}^{N}$ $\gamma_{nk}$ represents the the effective number of data points generated by $k^{th}$ Gaussian  </h4> </dd>


<dt> <h4>  3. Maximization step :  Update the parameters ($\hat\mu_{k}$ , $\hat\Sigma_{k}$ and $\hat\pi_{k}$) of GMM using the current resosibilities ($\gamma_{nk}$) by maximizing the likelihood.</h4> </dt>

<dd> <h4>
 - Effective number of data points generated by $k^{th}$ Gaussian : $N_{k}$ = $\sum_{n=1}^{N}$ $\gamma_{nk}$   
</h4> </dd>

<dd> <h4>
 - Mean of $k^{th}$ Gaussian : $\hat{\mu_{k}} = \frac{1}{N_{k}} \sum_{n=1}^{N}\gamma_{nk} x_{n}$   
</h4> </dd>

<dd> <h4>
 - Covariance of $k^{th}$ Gaussian : $\hat{\Sigma_{k}} = \frac{1}{N_{k}} \sum_{n=1}^{N} \gamma_{nk} ( x_{n} - \hat{\mu_{k}} )( x_{n} - \hat{\mu_{k}} )^{T}$   
</h4> </dd>

<dd> <h4>
 - Weight of $k^{th}$ Gaussian : $\hat{\pi_{k}} = \frac{N_{k}}{N}$  
</h4> </dd>

<dt> <h4> 4. Compute the log-likelihood with the updated parameters </h4> </dt>
<dt> <h4>  5. Repeat the expectation and maximization steps untill convergence </h4> </dt>

```python
#All imports
import numpy as np
import pandas as pd
import random as rand
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from scipy.stats import norm, multivariate_normal
import sys
import math

def generateData(pi, mu, sigma, N):
    ############################################################################
    #pi = weights of Gaussians, mu = mean of Gaussians, sigma = convariance of Gaussians, N = number of data points
    #Steps
    #1.Select th Gaussian following the probability distribution as pi
    #2.Sample the data points from the selected Gaussian
    #3.Repeat the process N times to get N data points
    #4.Return the sampled data
    ############################################################################
    #You may look into "random.choices" and "np.random.multivariate_normal" functions
    data = []
    labels = []
    for n in range(N):
      c = rand.choices([0,1], pi, k=1)
      m = c[0]
      s = np.random.multivariate_normal(mu[m],sigma[m], 1)
      labels.append(m)
      data.append(s[0])
    return np.asarray(data), np.asarray(labels)

    ############################################################################


############################################################################
#Chose the appropriate means ($\mu$), covariances ($\Sigma$), and weights
############################################################################
mu1 = [-1.0 , -1.0]
sig1 = [ [0.15 , 0.1], [0.1 , 0.25] ]

mu2 = [ 1.0, 1.0]
sig2 = [ [ 0.3, -0.25], [ -0.25, 0.25] ]

pi = [0.6 , 0.4]

n_points = 400

# Spherical covariances
sig1_s = [ [0.1 , 0.0], [0.0 , 0.2] ]
sig2_s = [ [ 0.2, 0.0], [ 0.0, 0.1] ]
# Diagonal covariances
sig1_d = [ [0.2 , 0.0], [0.0 , 0.2] ]
sig2_d = [ [ 0.1, 0.0], [ 0.0, 0.1] ]

############################################################################
#Create the synthetic data set
############################################################################

data_full, labels_full = generateData(pi, [mu1, mu2], [sig1, sig2], 200)

data_sp, labels_sp = generateData(pi, [mu1, mu2], [sig1_s, sig2_s], 200)
data_dg, labels_dg = generateData(pi, [mu1, mu2], [sig1_d, sig2_d], 200)

############################################################################
#Plot the synthetic data set
############################################################################


fig, ax = plt.subplots(1,3, figsize=(20,7))
x_min, x_max = data_full[:, 0].min() - 1, data_full[:, 0].max() + 1
y_min, y_max = data_full[:, 1].min() - 1, data_full[:, 1].max() + 1
cmap_bold = ListedColormap(['#FF0000', '#00FF00'])
ax[0].scatter(data_full[:, 0], data_full[:, 1], c=labels_full, cmap=cmap_bold)
ax[0].set_xlim(x_min, x_max)
ax[0].set_ylim(y_min, y_max)
ax[0].set_title('Full Covariance')

ax[1].scatter(data_sp[:, 0], data_sp[:, 1], c=labels_sp, cmap=cmap_bold)
ax[1].set_xlim(x_min, x_max)
ax[1].set_ylim(y_min, y_max)
ax[1].set_title('Spherical Covariance')

ax[2].scatter(data_dg[:, 0], data_dg[:, 1], c=labels_dg, cmap=cmap_bold)
ax[2].set_xlim(x_min, x_max)
ax[2].set_ylim(y_min, y_max)
ax[2].set_title('Diagonal Covariance')

plt.show()
```

```python
#All imports
import numpy as np
import pandas as pd
import random as rand
import matplotlib.pyplot as plt
from scipy.stats import norm, multivariate_normal
import sys
import math

############################################################################
#Write a function to compute the log-likelihood of the data given the parameters
############################################################################
def likelihood(data, pi, mu, sigma, log_likelihood=False, likelihoods=False):
  j = []
  for n in range(len(data)):
    s = 0
    for m in range(len(pi)):
      #print(np.linalg.det(sigma[m]))
      g = (1/np.sqrt(((2*np.pi)**len(data[n]))*np.linalg.det(sigma[m])))*np.exp((-1/2.)*(data[n]-mu[m])@np.linalg.inv(sigma[m])@(data[n]-mu[m]))
      s = s+g*pi[m]
    if log_likelihood:
      j.append(np.log10(s))
    else:
      j.append(s)
  if log_likelihood:
    return np.sum(j)
  if likelihoods:
    return j
  return np.prod(j)




############################################################################
#Write a function for the expectation step
############################################################################
def expectation(data, pi, mu, sigma):
  gamma = np.zeros((len(data), len(pi)))
  for n in range(len(data)):
    for k in range(len(pi)):
      numerator = likelihood([data[n]], [pi[k]], [mu[k]], [sigma[k]])
      denominator = likelihood([data[n]], pi, mu, sigma)
      gamma[n,k] = numerator/denominator
  return gamma







############################################################################
#Write a function for the maximization step.
############################################################################
def maximization(data, gamma):
  N_k = np.sum(gamma, axis=0)
  mu = np.zeros((gamma.shape[-1], data.shape[-1]))
  for k in range(len(mu)):
    mu_k = np.zeros((data.shape[-1]))
    for n in range(len(data)):
      mu_k = mu_k+gamma[n,k]*data[n]
    mu[k] = mu_k/N_k[k]
  sigma = np.zeros((gamma.shape[-1], data.shape[-1], data.shape[-1]))
  for k in range(len(mu)):
    sigma_k = np.zeros((data.shape[-1], data.shape[-1]))
    for n in range(len(data)):
      sigma_k = sigma_k+gamma[n,k]*(np.reshape(data[n]-mu[k], (-1,1))@np.reshape(data[n]-mu[k], (1,-1)))
    sigma[k] = sigma_k/N_k[k]
  pi = N_k/len(data)

  return pi, mu, sigma



def plot_contours(data, labels, pi, mu, sigma):
  w1 = np.linspace(-2.5,2.5,100)
  w0 = np.linspace(-2.5,2.5,100)
  W1, W0 = np.meshgrid(w1,w0)
  W = np.column_stack([np.ravel(W1), np.ravel(W0)])
  cmap_bold = ListedColormap(['#FF0000', '#00FF00'])
  plt.scatter(data[:, 0], data[:, 1], c=labels, cmap=cmap_bold)

  for k in range(len(pi)):
    likelihoods = likelihood(W, [pi[k]],[mu[k]],[sigma[k]], likelihoods=True)
    likelihoods = np.reshape(likelihoods, W0.shape)
    plt.contour(W0,W1,likelihoods)

  plt.show()



############################################################################
#Run the expectation and maximzation algorithm to estimate the parameters
############################################################################
data = data_full
labels = labels_full
N = data.shape[0] # number of input examples
D = data.shape[-1] # data dimension
M = 2 # no.of mixtures

pi = np.random.uniform(0.,3.,(M,))
pi = pi/np.sum(pi)

mu = np.random.uniform(-2.5,2.5,(M,D))
sigma = [np.diag(0.5*np.ones((D), dtype=np.float32)) for _ in range(M)]
#mu = [np.mean(data,axis=0)+2*m for m in range(M)]
#sigma = [np.cov(data.T) for _ in range(M)]



l = [-1000]
i = 0
patience = 3
count=0
while l[-1]<-50:
  ll = likelihood(data, pi, mu, sigma, log_likelihood=True)
  if ll<=l[-1]:
    count+=1
  else:
    count = 0
  if count>=patience:
    break
  print('Iteration:',i,'\t Log-likelihood',ll)
  plot_contours(data, labels, pi, mu, sigma)
  i+=1
  l.append(ll)
  gamma = expectation(data, pi, mu, sigma)
  pi, mu, sigma = maximization(data, gamma)



```

<h4> <b> Statistical approaches to speaker verification (GMM-UBM models) : </b>  This section describes the steps required to build an ASV system using GMM-UBM models.   </h4>

[link to data](https://drive.google.com/drive/folders/1szWD0-6pSSPAqaCCf80FZf0X2sofZ28w?usp=sharing)

<dt> <h4> 1. Feature Extraction </h4> </dt>
<dd> <h4> - Extract 30-dimensional Mel Frequency Cepstral Coefficients (MFCCs) from the speech signal. Add delta (velocity) and delta-delta (acceleration) coefficients to include the formant dynamics in the features. Use the following parameters for extracting the features. Framesize = 25 msec, Frameshift = 10 msec and number of mel filter banks = 40  </h4> </dd>
<dt> <h4> 2. Universal Background Modeling (UBM) </h4> </dt>
<dd> <h4> - Let us consider a Gaussian Mixture Model (GMM) with 128 Gaussians.  </h4> </dd>
<dd> <h4> - Estimate the parameters of the GMM model from large number of background speakers. Make sure that the background speakers are not involved in testing.   </h4> </dd>
<dd> <h4> - Speech data from 100 speakers are provided to estimate the UBM parameters. If you do not have sufficient computational resources, please use less number of speakers to estimate the parameters.   </h4> </dd>
<dt> <h4> 2. Speaker-specific GMM's </h4> </dt>
<dd> <h4> - Inputs : Trained UBM model i.e $\lambda^{ubm}$ and speaker training data i.e X   </h4> </dd>
<dd> <h4> - Stats estimation from UBM model: Compute the responsibility of the Gaussians with the initial estimates as UBM parameters </h4> </dd>
<dd> <h4> <center> $\gamma_{nk}$ = $\frac{\pi_{k}^{ubm} N ( x_{n} / \mu_{k}^{ubm} , \sigma_{k}^{2} )}{ \sum_{m=1}^{M} \pi_{m}^{ubm} N ( x_{n} / \mu_{m}^{ubm} , \sigma_{m}^{2} ) }$ </center> </h4> </dd>
<dd> <h4> where $\gamma_{nk}$ represents the responsibility of $k^{th}$ gaussian in generating $n^{th}$ data point and {$ \pi_{m}^{ubm} , \mu_{m}^{ubm} , \sigma_{m}^{2} $} are the UBM parameters    </h4> </dd>

<dd> <h4> - EM update for $\mu_{k}$ and  $\pi_{k}$: Estimate the parameters of the GMM { $\mu_{k}$,$\pi_{k}$ } using the available speaker's data  </h4> </dd>
<dd> <h4> <center> $\mu_{k}^{em}$ = $\frac{1}{N_{k}}$ $\sum_{n=1}^{N}$ $\gamma_{nk}x_{n}$ </center>
<center> $\pi_{k}^{em}$ = $\frac{\sum\limits_{n=1}^{N} \gamma_{nk}}{N} $ </center>  </h4> </dd>

<dd> <h4> - MAP update : Considering the UBM model as the prior information, The MAP adaption results in the following updates to zeroth order{$\pi_{k}$} and first-order statistics ($\mu_{k}$)   </h4> </dd>
<dd> <h4> <center> $\mu_{k}^{spk}$ = $\alpha\mu_{k}^{em}$ + $(1-\alpha)\mu_{k}^{ubm}$ </center> <br>
<center> $\pi_{k}^{spk}$ = $\alpha\pi_{k}^{em}$ + $(1-\alpha)\pi_{k}^{ubm}$ </center>   </h4> </dd>
<dd> <h4>  Where $\alpha$ is given by $\frac{N_{k}}{N_{k} + r}$ and "r" is the relevance factor.  </h4> </dd>

<dt> <h4> 3. Compute the likelihood ratio </h4> </dt>
<dd> <h4> - Estimate the score for the hypothesis that the given speaker came from the hypothesized speaker  </h4> </dd>
<dd> <h4> - Let X be the set of feature vectors from the test utterance  </h4> </dd>
<dd> <h4> - Let k be the claimed identity of the speaker.
<dd> <h4> - H0 = Feature set X is from speaker k  </h4> </dd>
<dd> <h4> - H1 = Feature set X is not from speaker k   </h4> </dd>
<dd> <h4> - Hypothesis testing based on likelihood ratio   </h4> </dd>

<dd> <h4>
</br>
<center>  $\frac{P(X/\lambda_{k})}{P(X/\lambda_{\bar{k}})}$ >= $\theta$ accepts H0 </center> </br>
<center>  $\frac{P(X/\lambda_{k})}{P(X/\lambda_{\bar{k}})}$ < $\theta$ reject H0 </center>  <br>
</h4> </dd>

<dd> <h4> $\lambda_{k}$ denotes model estimated from reference utterance of k. </h4> </dd>
<dd> <h4> $\lambda_{\bar{k}}$ denotes model estimated from entire world except k. Let us consider UBM model here. </h4> </dd> </h4> </dd>

```python
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
import python_speech_features as pfeatures
import os
from sklearn.mixture import GaussianMixture
import pickle

baseDir = '/content/drive/MyDrive/TalentSprint/Session1/SpeakerVerificationData/'
UBMDir = baseDir + "UBM_data/"
enrollDir = baseDir + "Enrollment/"
testDir = baseDir + "Test/"

#Function to compute MFCC features of single file
def getFeatures(filename):
    (sig,rate) = sf.read(filename)
    mfcc = pfeatures.mfcc(sig, rate, winlen=0.025, winstep=0.01, numcep = 23, nfilt=26,
    preemph=0.97, appendEnergy=True)
    d_mfcc = pfeatures.delta(mfcc, 2)
    a_mfcc = pfeatures.delta(d_mfcc, 2)
    out = np.concatenate([mfcc, d_mfcc, a_mfcc], axis=1)
    return out

#Function to compute MFCC features of entire files in the folder
def spkfeats(folder,curDir):
    #print("Okay",folder)
    wav_list = []
    for path, subdirs, files in os.walk(curDir + folder):
        for name in files:
            wav_list.append(os.path.join(path, name))
    feats=[]
    for wav in wav_list:
        feats.append(getFeatures(wav))
    return feats



##############################
#Building UBM model
##############################
ubm_spk = os.listdir(UBMDir)
for i in range(0,10,1): #len(ubm_spk)
    if i == 0:
        UBMfeats = np.vstack(spkfeats(ubm_spk[i],UBMDir))
    else:
        cfeats = np.vstack(spkfeats(ubm_spk[i],UBMDir))
        UBMfeats = np.vstack((UBMfeats,cfeats))
sampIndexes = np.random.choice(len(UBMfeats),500000) #subsampling the frames
sampUBMfeats = UBMfeats[sampIndexes]
UBMModel = GaussianMixture(n_components=24,max_iter=100000,covariance_type='diag',verbose=2)
UBMModel.fit(sampUBMfeats)


##############################
#Building speaker-specific GMM models
##############################


enrollSpks = os.listdir(enrollDir)
testSpks = os.listdir(testDir)
enrollSpks_feats={}
enrollSpks_models={}


relevance_factor = 0.9

def likelihood(data, pi, mu, sigma, log_likelihood=False, likelihoods=False):
  j = []
  for n in range(len(data)):
    s = 0
    for m in range(len(pi)):
      #print(np.linalg.det(sigma[m]))
      g = (1/np.sqrt(((2*np.pi)**len(data[n]))*np.linalg.det(sigma[m])))*np.exp((-1/2.)*(data[n]-mu[m])@np.linalg.inv(sigma[m])@(data[n]-mu[m]))
      s = s+g*pi[m]
    if log_likelihood:
      j.append(np.log10(s))
    else:
      j.append(s)
  if log_likelihood:
    return np.mean(j)
  if likelihoods:
    return j
  return np.prod(j)

def adaptMAP(UBMModel,spkfeatures,relevance_factor=16):
    labels = UBMModel.predict_proba(enrollSpks_feats[enrollSpks[i]])
    N_k = np.sum(labels.T, axis=1, keepdims=True)
    alpha = N_k/(N_k+relevance_factor)
    means_em = (labels.T@ enrollSpks_feats[enrollSpks[i]])/np.sum(labels.T,axis=1, keepdims=True)
    weights_em = np.sum(labels,axis=0)/np.sum(labels)

    weights_map = (1-alpha)*UBMModel.weights_ + alpha*weights_em
    means_map = (1-alpha)*UBMModel.means_ + alpha*means_em
    convariances_map = UBMModel.covariances_

    return weights_map,means_map,convariances_map

for i in range(0,len(enrollSpks),1):
    enrollSpks_feats[enrollSpks[i]]=np.vstack(spkfeats(enrollSpks[i],enrollDir))
    spk_weights, spk_means, spk_covariances = adaptMAP(UBMModel,enrollSpks_feats[enrollSpks[i]])
    enrollSpks_models[enrollSpks[i]] = [spk_weights, spk_means, spk_covariances]

testSpk = testDir+'Speaker1/0LbtndiXJC0/00010.wav'

likelihood_ratios = {}
testSpk_feats = getFeatures(testSpk)
if len(UBMModel.covariances_.shape) == 2:
  ubm_covariances = [np.diag(UBMModel.covariances_[i]) for i in range(len(UBMModel.covariances_))]
  ubm_covariances = np.asarray(ubm_covariances)
ubm_likelihood = likelihood(testSpk_feats, UBMModel.weights_, UBMModel.means_, ubm_covariances, log_likelihood=True)
for enr_spkr, enr_model in enrollSpks_models.items():
  if len(enr_model[2].shape) == 2:
    enr_model[2] = [np.diag(enr_model[2][i]) for i in range(len(enr_model[2]))]
    enr_model[2] = np.asarray(enr_model[2])
  log_likelihood = likelihood(testSpk_feats, enr_model[0], enr_model[1], enr_model[2], log_likelihood=True)
  likelihood_ratios[enr_spkr] = log_likelihood-ubm_likelihood
print("Identified speaker = ",max(likelihood_ratios, key=likelihood_ratios.get))
```
